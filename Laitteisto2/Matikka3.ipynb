```python
#Tehtävä 2.1

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = x**3
y3 = x**4




ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_ylim([-10, 10])
ax.set_xlim([-6, 6])
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```
