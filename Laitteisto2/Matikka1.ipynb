```python
#Tehtävä 1
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 3)
y = np.where((x >= 1) & (x <= 3), -2*x + 6, np.where((x >= -1) & (x <= 1), 4, np.nan))
ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_aspect('equal')
ax.set_ylim([-6, 6])
ax.set_xlim([-6, 6])
plt.plot(x, y)

```

```python
#Tehtävä 2

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 10)
y = np.where((x >= -5) & (x <= 0), 3/5*x+3, np.where((x >= 0) & (x <= 4), 3,np.where((x >= 4) & (x <= 8), -3/4*x+6,np.where((x >= 8) & (x <= 10), 1/2*x-4,np.nan))))

ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_ylim([-6, 6])
ax.set_xlim([-6, 6])

plt.plot(x, y)
plt.show
```
